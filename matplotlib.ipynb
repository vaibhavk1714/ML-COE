```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [7, 16, 24, 27, 32]
```

```python
plt.plot(x, y, 'r--', label = 'Dashed Line Plot')
plt.legend()
```

```python
plt.subplot(1, 1, 1)
plt.title('Bar Graph')
plt.bar(x, y, color = 'blue')
```

```python
plt.subplot(2, 1, 2)
plt.hist(y, bins = 5, color = 'green', edgecolor = 'black')
plt.title('Histogram')
```

```python
plt.subplot(2, 2, 2)
plt.pie(y, labels = x, autopct='%1.1f%%', startangle = 90)
plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```
